```python
from selenium import webdriver
import time
import pandas as pd
```

```python
data = pd.DataFrame(columns=['Date','Asset','Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])
etfs = ['VTI','VXUS','BND','BNDX']
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')

for etf in etfs:
    asset = etf
    url = 'https://finance.yahoo.com/quote/'+asset+'/history'
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    driver.find_element_by_xpath("//div[@class='Pos(r) D(ib) C($linkColor) O(n):f Cur(p)']//*[name()='svg']").click()
    driver.find_element_by_xpath("//div[@class='W(195px)']//ul[2]//li[4]//button[@class = 'Py(5px) W(45px) Fz(s) C($tertiaryColor) Cur(p) Bd Bdc($seperatorColor) Bgc($lv4BgColor) Bdc($linkColor):h Bdrs(3px)']").click()
    driver.find_element_by_xpath("//div[@class='Bgc($lv1BgColor) Bdrs(3px) P(10px)']//button[1]").click()
    x = driver.find_element_by_xpath("//div[@class='C($tertiaryColor) Mt(20px) Mb(15px)']//a[1]").get_attribute('href')
    driver.close()
    df = pd.read_csv(x)
    df['Asset'] = asset
    data = data.append(df)
    print(asset)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```
